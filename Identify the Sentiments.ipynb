Problem Statement

Sentiment analysis remains one of the key problems that has seen extensive application of natural language processing. This time around, given the tweets from customers about various tech firms who manufacture and sell mobiles, computers, laptops, etc, the task is to identify if the tweets have a negative sentiment towards such companies or products.

 

Evaluation Metric
The metric used for evaluating the performance of classification model would be weighted F1-Score.

 

Public and Private Split
Note that the test data is further randomly divided into Public (35%) and Private (65%) data. Your initial responses will be checked and scored on the Public data. The final rankings would be based on your private score which will be published once the competition is over.

## Preprocessing

### Removing Links

```python
train['tweet']=train['tweet'].str.replace('http\S*|www\S*',"",case=False)
test['tweet']=test['tweet'].str.replace('http\S*|www\S*',"",case=False)
```

### Removing Emojis

```python
def remove_emoji(tweet):
    emoji_pattern=re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',tweet)

train['tweet']=train['tweet'].apply(remove_emoji)
test['tweet']=test['tweet'].apply(remove_emoji)
```

### Removing Numbers

```python
def remove_numbers(tweet):
    tweet=''.join([i for i in tweet if not i.isdigit()])         
    return tweet

train['tweet']=train['tweet'].apply(remove_numbers)
test['tweet']=test['tweet'].apply(remove_numbers)
```

### Removing Twitter Handles

```python
train['tweet']=train['tweet'].str.replace("@[\w]*","")
test['tweet']=test['tweet'].str.replace("@[\w]*","")
```

### Removing Special Characters

```python
train['tweet']=train['tweet'].str.replace("[^a-zA-Z]"," ")
test['tweet']=test['tweet'].str.replace("[^a-zA-Z]"," ")
```

### Removing Repeating Words

```python
train['tweet']=train['tweet'].apply(lambda x : re.sub(r'(.)\1{1,}', r'\1\1',x))
test['tweet']=test['tweet'].apply(lambda x : re.sub(r'(.)\1{1,}', r'\1\1',x))
```

### Removing Accented Characters

```python
def remove_accented_chars(tweet):
    tweet=unicodedata.normalize('NFKD',tweet).encode('ascii','ignore').decode('utf-8','ignore')
    return tweet

train['tweet']=train['tweet'].apply(remove_accented_chars)
test['tweet']=test['tweet'].apply(remove_accented_chars)
```

```python
train['tweet']=train['tweet'].str.lower()
test['tweet']=test['tweet'].str.lower()
```

```python
def expand_contractions(tweet,contraction_mapping=CONTRACTION_MAP):
    
    contractions_pattern=re.compile('({})'.format('|'.join(contraction_mapping.keys())), 
                                      flags=re.IGNORECASE|re.DOTALL)
    def expand_match(contraction):
        match=contraction.group(0)
        first_char=match[0]
        expanded_contraction = contraction_mapping.get(match)\
                                if contraction_mapping.get(match)\
                                else contraction_mapping.get(match.lower())                       
        expanded_contraction = first_char+expanded_contraction[1:]
        return expanded_contraction
        
    expanded_tweet=contractions_pattern.sub(expand_match,tweet)
    expanded_tweet=re.sub("'","",expanded_tweet)
    return expanded_tweet

train['tweet']=train['tweet'].apply(expand_contractions)
test['tweet']=test['tweet'].apply(expand_contractions)
```

### Removing Stopwords

```python
tokenizer=ToktokTokenizer()
stopword_list=nltk.corpus.stopwords.words('english')

def remove_stopwords(tweet,is_lower_case=False):
    tokens=tokenizer.tokenize(tweet)
    tokens=[token.strip() for token in tokens]
    filtered_tokens=[token for token in tokens if token not in stopword_list]
    filtered_tweet= ' '.join(filtered_tokens)    
    return filtered_tweet

train['tweet']=train['tweet'].apply(remove_stopwords)
test['tweet']=test['tweet'].apply(remove_stopwords)
```

### Lemmatization

 Lemmatization is the process of grouping together the different inflected forms of a word so they can be analysed as a single item. Lemmatization is similar to Stemming but it brings context to the words. So it links words with similar meaning to one word.

Text preprocessing includes both Stemming as well as Lemmatization. Actually, Lemmatization is preferred over Stemming because Lemmatization does morphological analysis of the words.

```python
# splitting the text (tokenizing)
tokenized_data_train=train['tweet'].apply(lambda x: x.split()) 
tokenized_data_test=test['tweet'].apply(lambda x: x.split()) 

lemmatizer=WordNetLemmatizer()

# lemmatizing the tokenized text
tokenized_data_train=tokenized_data_train.apply(lambda x: [lemmatizer.lemmatize(i) for i in x])
tokenized_data_test=tokenized_data_test.apply(lambda x: [lemmatizer.lemmatize(i) for i in x])

# again joining the tokenized text
for i in range(len(tokenized_data_train)):
    tokenized_data_train[i]=' '.join(tokenized_data_train[i])

for i in range(len(tokenized_data_test)):
    tokenized_data_test[i]=' '.join(tokenized_data_test[i])    

# saving the clean text    
train['tweet']=tokenized_data_train
test['tweet']=tokenized_data_test 
```

# Train - Validation Split

```python
X_train,X_val,y_train,y_val=train_test_split(train.tweet,
                                             train.label,
                                             test_size=0.30,
                                             stratify=train.label,
                                             random_state=10)
```

# TF - IDF
In information retrieval, TFIDF, short for term frequency – inverse document frequency, is a numerical statistic that is intended to reflect how important a word is to a document in a collection or corpus.

```python
tv=TfidfVectorizer()

#transformed train set
tv_train=tv.fit_transform(X_train)

#transformed validation set
tv_val=tv.transform(X_val)

#transformed test set
tv_test=tv.transform(test.tweet)

print('tfidf_train:',tv_train.shape)
print('tfidf_validation:',tv_val.shape)
print('tfidf_test:',tv_test.shape)
```

# Model Building

## Logistic Regression

```python
# defining classifier
lr=LogisticRegression()

# fitting for tfidf vectorizer.
tfidf_lr=lr.fit(tv_train,
                y_train)

# predicting and printing f1 score for train and validation data
tfidf_train_predict_lr=tfidf_lr.predict(tv_train)
tfidf_val_predict_lr=tfidf_lr.predict(tv_val)

f1_lr_tr=f1_score(y_train,
                  tfidf_train_predict_lr,
                  average='weighted')
f1_lr_val=f1_score(y_val,
                   tfidf_val_predict_lr,
                   average='weighted')
print('Train F1 Score')
print(f1_lr_tr)
print('Val F1 Score')
print(f1_lr_val)
```

```python
prediction_lr=lr.predict(tv_test)

submission=pd.DataFrame({"id":test["id"],
                         "label":prediction_lr})
submission.to_csv('sub_lr.csv',index=False)
```
